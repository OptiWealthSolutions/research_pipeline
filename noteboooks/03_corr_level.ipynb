```python
import numpy as np
import yfinance as yf
import pandas_datareader.data as web
import pandas as pd
import matplotlib.pyplot as plt
import os
import fredapi as fred
import statsmodels.api as sm
start_date = "2010-01-01"
end_date = "2025-01-01"
```

```python
def fetchPrice(tickers):
    data = yf.download(tickers,start=start_date,end=end_date,interval="1d")['Close']
    data[f"{tickers[0]}_change"] = data[tickers[0]].pct_change()
    data[f"{tickers[1]}_change"] = data[tickers[1]].pct_change()
    data["roll_corr"] = data[f"{tickers[0]}_change"].rolling(window=60).corr(
        data[f"{tickers[1]}_change"]
    )
    return data
```

```python
def run_multiple_linear_regression(data):
    data = data.dropna()
    y = data[f"{tickers[0]}_change"]
    X = data.drop(columns=f"{tickers[0]}_change")
    X = sm.add_constant(X)
    # Affichage du rapport

    model = sm.OLS(y,X).fit()

    print(model.summary())

    return model

tickers = ["EURUSD=X", "GBPUSD=X"]
data = fetchPrice(tickers)

independent_variables = ["EURUSD=X_change", "roll_corr"]


regression_model = run_multiple_linear_regression(
    data
)

```

```python

#run graph
def plot_simple_regression(model, X, y):
    predictions = model.predict(X)
    plt.figure(figsize=(8, 6))
    plt.scatter(y, predictions, alpha=0.7)
    plt.show()
    return predictions
plot_simple_regression(regression_model,data.drop(columns=f"{tickers[0]}_change"),data[f"{tickers[0]}_change"])
```
